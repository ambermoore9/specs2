### Step 2: Treat Names / Casing / Spelling Issues

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
# Drop "Unnamed: 0" column from DataFrame

### CLEAN ### 
house_df.drop(columns=['Unnamed: 0'], inplace=True)

### TEST ###
house_df.head()
```

```python
### DEFINE ###
# same thing as above but for other database

### CLEAN ### 
sales_df.drop(columns=['Unnamed: 0'], inplace=True)

### TEST ###
sales_df.head()
```

```python
### DEFINE ###
# columns lowercase in house database

### CLEAN ### 
house_df.columns = house_df.columns.str.lower()

### TEST ###
house_df.head()
```

```python
### DEFINE ###
# columns lowercase in sales database

### CLEAN ### 
sales_df.columns = sales_df.columns.str.lower()

### TEST ###
sales_df.head()
```

```python
### DEFINE ###
# remove periods and spaces and snake case

### CLEAN ### 
import regex as re
column_list = house_df.columns.to_list()
new_columns = []
for column in column_list:
    column = column.replace(" ", "_")
    column = column.replace("?", "")
    new_columns.append(column)
house_df.columns = new_columns

### TEST ###
house_df.head()
```

```python
### DEFINE ###
#change data type house

### CLEAN ### 
house_df.id = house_df.id.astype(str)

### TEST ###
print(house_df['id']) 
```

```python
### DEFINE ###
#change data type sales

### CLEAN ### 
sales_df.id = house_df.id.astype(str)

### TEST ###
print(sales_df.id)
```

```python
### DEFINE ###
#change zipcode in house database

### CLEAN ### 
house_df['zipcode'] = house_df['zipcode'].astype(str)
house_df['zipcode'] = house_df['zipcode'].map(lambda x: x.rstrip('.0'))

### TEST ###
print(house_df.zipcode)
```

```python
### DEFINE ###
# date column to datetime

### CLEAN ### 
if sales_df['date'].dtype == 'O':
    sales_df['date'] = sales_df['date'].str.replace('T000000', '')

sales_df['date'] = pd.to_datetime(sales_df['date'], format='%Y%m%d')

### TEST ###
print(sales_df['date'])
```

```python
### DEFINE ###
# price column float

### CLEAN ### 
sales_df['price'] = sales_df['price'].astype(str).str.replace('$', '', regex=False)
sales_df['price'] = sales_df['price'].str.replace(',', '', regex=False).astype(float)

### TEST ###
print(sales_df['price'])
```

### Step 3: Treat Missing Values

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
# Drop the Year Renovated column due to majority of data missing

### CLEAN ###  
house_df.drop(columns=['year_renovated'], inplace=True)

### TEST ###
house_df.info()
```

```python
### DEFINE ###
# fill bathroom column with means

### CLEAN ### 
house_df.bathrooms = house_df.bathrooms.fillna(house_df.bathrooms.mean())

### TEST ###
print(house_df.bathrooms.isnull().sum())
```

```python
### DEFINE ###
# fill grade column with mean

### CLEAN ### 
house_df['grade'] = house_df['grade'].fillna(value=house_df['grade'].mean())

### TEST ###
house_df['grade'].isnull().sum() 
```

```python
### DEFINE ###
# fill condition column with mode

### CLEAN ### 
house_df['house_condition'].fillna(house_df['house_condition'].mode()[0], inplace=True)

### TEST ###
print(house_df['house_condition'])
```

```python
### DEFINE ###
# drop rows in the dataframe with null values

### CLEAN ### 
house_df.dropna(axis=0, how='any', inplace=True)

### TEST ###
house_df.isnull().sum()
```

```python
### DEFINE ###
# delete null rows in price sales database

### CLEAN ### 
sales_df.dropna(axis=0, how='any', inplace=True)

### TEST ###
sales_df.isnull().sum()
```
